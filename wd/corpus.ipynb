```python
from lib.wikidata import query_wikidata

SPARQL_QUERY = """
SELECT *  {
  ?item wdt:P361 wd:Q137383524 .
  #OPTIONAL {?item wdt:P50 ?author.}
  OPTIONAL {?item wdt:P577 ?date.}
  OPTIONAL {?item wdt:P291 ?place.}
  OPTIONAL {?item p:P361/pq:P217 ?index.}
  MINUS {?item wdt:P50 wd:Q245369.}
}
"""

source = "S248\tQ137383524\tS813\t+2026-03-23T00:00:00Z/11"

results = query_wikidata(SPARQL_QUERY)
for result in results["results"]["bindings"]:
    item = result["item"]["value"][31:]
    date = result.get("date") or None
    if date is not None:
        date = date["value"]
    place = result.get("place") or None
    if place is not None:
        place = place["value"][31:]
    author = result.get("author") or None
    if author is not None:
        author = author["value"][31:]

    if date is not None:
        print(f"{item}\tP577\t+{date}/9\t{source}")
    if place is not None:
        print(f"{item}\tP291\t{place}\t{source}")
    print(f"{item}\tP7937\tQ8261\t{source}")
    if author is not None:
        print(f"{item}\tP50\t{author}\t{source}")
```

```python
import pandas as pd
from lib.wikidata import query_wikidata

SPARQL_QUERY = """
SELECT ?index ?date {
  ?item wdt:P361 wd:Q137383524 .
  OPTIONAL {?item wdt:P577 ?date.}
  OPTIONAL {?item p:P361/pq:P217 ?index.}
  OPTIONAL {}
}
"""

results = query_wikidata(SPARQL_QUERY)
results = pd.read_sql_query=results["results"]["bindings"]
flipped = {}
for result in results:
    if result["index"]["value"] not in flipped:
        flipped[result["index"]["value"]] = 1
    else:
        flipped[result["index"]["value"]] += 1
print("\n".join(f"{k}\t{v}" for k, v in flipped.items() if v > 1))
```
